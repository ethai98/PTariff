```python
import pandas as pd
import re
import os

# Define input and output directories
input_dir = "/Users/ericthai/Library/CloudStorage/Dropbox/1. Research/3. Data/PTariff/2.tabula_processed_schedule/EVL_remaining"     # Replace this with your actual input folder path
output_dir = "/Users/ericthai/Library/CloudStorage/Dropbox/1. Research/3. Data/PTariff/3.gpt_processed_schedule/python_processed_EVL_remaining"   # Replace this with your actual output folder path

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)
```

```python
def clean_tariff_file(filepath, output_dir):
    df = pd.read_csv(filepath, dtype=str)

    # Standardize column names
    df.columns = df.columns.str.strip().str.lower().str.replace('\r', '', regex=False)

    # Rename relevant columns
    column_mapping = {
        col: "code" for col in df.columns if "code" in col.lower() or "tariff reference" in col.lower()
    }
    column_mapping.update({
        col: "baserate" for col in df.columns if "baserate" in col.lower() or "base rate" in col.lower()
    })
    column_mapping.update({
        col: "category" for col in df.columns if "staging category" in col.lower() or "category" in col.lower()
    })
    df = df.rename(columns=column_mapping)

    # Keep only relevant columns
    if not all(col in df.columns for col in ["code", "baserate", "category"]):
        return  # Skip files missing essential columns

    df = df[["code", "baserate", "category"]]

    # Drop rows with no category
    df = df[df["category"].notna() & (df["category"].str.strip() != "")]

    # Remove carriage return characters
    df = df.replace({r'\r': ''}, regex=True)

    # Ensure all target columns are strings
    df["code"] = df["code"].astype(str)
    df["baserate"] = df["baserate"].astype(str)
    df["category"] = df["category"].astype(str)

    # Remove "." in code
    df["code"] = df["code"].str.replace(".", "", regex=False)

    # Keep codes with digits and optionally one trailing uppercase letter
    df = df[df["code"].str.match(r"^\d+[A-Z]?$", na=False)]

    # Drop rows with only 4-digit codes
    df = df[~df["code"].str.match(r"^\d{4}$")]

    # Clean baserate column
    df["baserate"] = df["baserate"].str.strip().str.lower()
    df["baserate"] = df["baserate"].replace({"free": "0", "libre": "0"})
    df["baserate"] = df["baserate"].str.replace("%", "", regex=False)
    df["baserate"] = df["baserate"].apply(lambda x: "PR" if re.search(r"[a-zA-Z]", x) and not x.isnumeric() else x)
    df["baserate"] = df["baserate"].replace("", "NA")

    # Drop rows with missing or blank values
    df = df.dropna()
    df = df[(df["code"].str.strip() != "") & (df["baserate"].str.strip() != "") & (df["category"].str.strip() != "")]

    # Save cleaned file
    filename = os.path.basename(filepath)
    cleaned_name = filename.replace("cleaned_", "").replace(".csv", "_gpt.csv")
    output_path = os.path.join(output_dir, cleaned_name)
    df.to_csv(output_path, index=False)
```

```python
for file_name in os.listdir(input_dir):
    if file_name.endswith(".csv"):
        file_path = os.path.join(input_dir, file_name)
        clean_tariff_file(file_path, output_dir)
```
