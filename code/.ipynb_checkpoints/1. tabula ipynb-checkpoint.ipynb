```python
import tabula
import os
```

Convert all PDFs files in directory into CSV tables (CHANGE PDF and CSV DIRECTORY for each batch)

```python
pdf_directory = "/Users/ericthai/Library/CloudStorage/Dropbox/1. Research/3. Data/PTariff/1.raw_schedules/EVL_remaining"

csv_directory = "/Users/ericthai/Library/CloudStorage/Dropbox/1. Research/3. Data/PTariff/2.tabula_processed_schedule/EVL_remaining"
```

```python
for filename in os.listdir(pdf_directory):
    if filename.endswith(".pdf"):
        pdf_file_path = os.path.join(pdf_directory, filename)
        csv_base_filename = os.path.splitext(filename)[0]
        csv_file_path = os.path.join(csv_directory, f"{csv_base_filename}.csv")
        tabula.convert_into(pdf_file_path, csv_file_path, output_format="csv", pages="all")
```
